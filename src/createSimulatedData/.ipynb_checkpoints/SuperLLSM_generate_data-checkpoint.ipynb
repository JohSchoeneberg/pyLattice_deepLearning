TRIVIAL APPROACH: Generate particle trajectories and microscopy images
----------
**Simulation setup / Ground truth**: 
* xy Area of 100 x 100 micrometer. 
* Typical diffusion constant of proteins on a membrane: 0.25 micrometer^2 / s. (In cytosol about 100 um^2/s)
* timestep 1s, 100 frames.
* 1000 particles.

**Input data**
* discretize to 100 nm grid.

```python
def simulate_particles(N=1000, size=10.0, timesteps=100, D=0.25, Damp=0.01, kamp=1):
    """ Simulate particles
    
    Parameters
    ----------
    N : number of particles
    size : x/y size in micrometers
    timesteps : number of timesteps in s
    D : diffusion constant in micrometer/2 / s
    
    """
    X = np.zeros((timesteps, N, 3))  # time, particle-index, xy
    A = np.zeros((timesteps, N))
    # initialize
    X[0] = size * np.random.rand(N, 3)
    # simulate
    for i in range(1, timesteps):
        # update positions
        X[i] = X[i-1] + D * np.random.randn(N, 3)
        # constrain to bounding box
        X[i] = X[i] % size
        # update amplitudes
        A[i] = A[i-1] -kamp*A[i] + Damp * np.random.randn(N)
        # constrain to [0, 1]
        A[i] = np.maximum(np.minimum(A[i], 1.0), 0.0)
    
    return X, A
```

```python
# relationship between resolution (FWHM) and sigma
standard_resolution = 0.100
standard_sigma = standard_resolution / (2.0 * np.sqrt(2*np.log(2)))
```

```python
def microscope(trajs_X, trajs_A, size, 
               gridsize=0.1, psf_width=standard_sigma, noise_mean=0.0, noise_amp=0.1, timestep=1):
    """ Simulated a series of microscopy images given a set of particle trajectories
    
    Parameters
    ----------
    trajs : numpy array (T x N x 2)
        T timesteps of N particle in 2D
    size : float
        Microscope image size, in micrometers
    gridsize : float
        xy-Size of one microscope pixel, in micrometers
    psf_width : float
        width of the point spread function, in micrometers
    noise : float
        Shot noise intensity
    timestep : int
        The number of steps between two images, for coarse-graining in time.
    
    """
    # coarse-grain time first
    trajs_X = trajs_X[::timestep]
    trajs_A = trajs_A[::timestep]
    
    # make grid
    timesteps, N, _ = trajs_X.shape
    
    ngrid = int(size / gridsize)
    c_grid = np.linspace(0, size-gridsize, num=ngrid) + 0.5*gridsize
    n_neighbors = int(3 * psf_width / gridsize) + 1
    X = np.zeros((timesteps, ngrid, ngrid, ngrid))
    
    # spread on grid
    for t in range(timesteps):
        #print(t)
        for i in range(N):
            # find central grid cell
            ix = int((trajs_X[t, i, 0] / size) * ngrid)
            iy = int((trajs_X[t, i, 1] / size) * ngrid)
            iz = int((trajs_X[t, i, 2] / size) * ngrid)
            # iterate over valid neighbors
            #print(i)
            for kx in range(max(0, ix - n_neighbors), min(ngrid, ix + n_neighbors)):
                for ky in range(max(0, iy - n_neighbors), min(ngrid, iy + n_neighbors)):
                    for kz in range(max(0, iz - n_neighbors), min(ngrid, iz + n_neighbors)):

                        d2 = (trajs_X[t, i, 0] - c_grid[kx])**2 + (trajs_X[t, i, 1] - c_grid[ky])**2 + (trajs_X[t, i, 2] - c_grid[kz])**2
                        value = trajs_A[t, i] * np.exp(-d2 / (2*psf_width*psf_width))
                        
                        X[t, kx, ky, kz] += value
                    
    # add noise
    noise = noise_mean + noise_amp * np.random.randn(timesteps, ngrid, ngrid, ngrid)
    noise = np.maximum(np.minimum(noise, 1.0), 0.0)
    X += noise
    return X
```

```python
imageSize = 2
trajs_X, trajs_A = simulate_particles(N=15, size=imageSize, timesteps=5, D=0.025)
X = microscope(trajs_X, trajs_A, size=imageSize, 
               gridsize=0.1, psf_width=standard_sigma, noise_mean=0.0, noise_amp=0.0, timestep=1)
```

```python
import skimage
import os
outputPath_tiff_prefix = './test__X_'
# this multiplier is necessary so that to prevent all the values being 0 when converting to int

multiplierForOutput = 2**16 #16 bit uint is the max value, so multiply all by that (otherwise you get 0 everywhere when converting to int)
for i in range(len(X)):
    outputPath_tiff = outputPath_tiff_prefix+str(i).zfill(2)+'.tiff'
    skimage.external.tifffile.imsave(outputPath_tiff_prefix+str(i).zfill(2)+'.tiff', (multiplierForOutput*X[i]).astype('uint16'))    
    print(os.path.abspath(outputPath_tiff))
    
np.sort(multiplierForOutput*X[1].flatten())

#open the result in chimera X
```
