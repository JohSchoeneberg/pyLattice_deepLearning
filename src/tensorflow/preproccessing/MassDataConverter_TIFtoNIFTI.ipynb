```python
import os
from skimage import io
import numpy
```

```python
image_directory="/home/gautham/bkly/LatticeLightSheet/TrialData/"
save_directory="/home/gautham/bkly/LatticeLightSheet/ConvertedData/"
```

```python
import nibabel as nib
import numpy as np

for filename in os.listdir(image_directory):
    if filename.endswith(".tif"): 
        #Convert TIF to numpy array
        image_path = os.path.join(image_directory, filename)
        im = io.imread(image_path)
        #Convert numpy array to NIFTI
        nifti_file = nib.Nifti1Image(im, affine=np.eye(4))
        print(type(nifti_file))
        nib.save(nifti_file, save_directory+filename[:-4]+'.nii.gz')
        continue
    else:
        print(filename+" not converted")
        continue
```
